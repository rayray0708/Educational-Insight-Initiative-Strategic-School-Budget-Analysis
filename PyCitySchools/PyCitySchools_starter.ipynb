```python
# Dependencies and Setup
import pandas as pd
from pathlib import Path

# File to Load (Remember to Change These)
school_data_to_load = Path("Resources/schools_complete.csv")
student_data_to_load = Path("Resources/students_complete.csv")

# Read School and Student Data File and store into Pandas DataFrames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset.  
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
school_data_complete
```

## Local Government Area Summary

* Calculate the total number of schools

* Calculate the total number of students

* Calculate the total budget

* Calculate the average maths score 

* Calculate the average reading score

* Calculate the percentage of students with a passing maths score (50 or greater)

* Calculate the percentage of students with a passing reading score (50 or greater)

* Calculate the percentage of students who passed maths **and** reading (% Overall Passing)

* Create a dataframe to hold the above results

* Optional: give the displayed data cleaner formatting

```python
#calculate number of schools first by using the unique function, and len() function
total_schools = len(school_data_complete["school_name"].unique())
# total_schools

#calculate the number of students
total_students = school_data_complete["student_name"].count()
# total_students

#sum all the unique budgets 
total_budget = school_data_complete["budget"].unique().sum()
# total_budget

#find mean of maths score
average_maths = school_data_complete["maths_score"].mean()
# find average reading score across all schools
average_reading = school_data_complete["reading_score"].mean()
# average_reading


# Calculate the percentage of students with a passing maths score (50 or greater)
## Use loc to filter out students who achieved lower than 50,
##to perform a function on this condition, put it inside round brackets
passed_maths = (school_data_complete["maths_score"] >= 50).sum() 

## Divide sum of those who passed maths by total number of students
percentage_passed_maths = passed_maths/total_students*100
percentage_passed_maths

# Calculate the percentage of students with a passing reading score (50 or greater)
## Use loc to filter out students who achieved lower than 50,
##to perform a function on this condition, put it inside round brackets
passed_reading = (school_data_complete["reading_score"] >= 50).sum()
passed_reading

## Divide sum of those who passed reading by total number of students
percentage_passed_reading = passed_reading/total_students*100
percentage_passed_reading

# Calculate the percentage of students who passed maths and reading (% Overall Passing)
passed_overall = len(school_data_complete.loc[(school_data_complete["maths_score"] >= 50)
                                          &(school_data_complete["reading_score"] >= 50),"student_name"])
passed_overall

## Divide sum of those who passed both maths AND reading by total number of students
percentage_passed_overall = passed_overall/total_students*100
percentage_passed_overall
                                           
# Create a dataframe to hold the above results

area_summary = pd.DataFrame({"Total Schools": [total_schools],
                                    "Total Students": total_students,
                                    "Total Budget": total_budget,
                                    "Average Maths Score": average_maths,
                                    "Average Reading Score": average_reading,
                                     "% Passing Maths": percentage_passed_maths,
                                     "% Passing Reading": percentage_passed_reading,
                                     "% Overall Passing": percentage_passed_overall})
area_summary


                            
```

## School Summary

* Create an overview table that summarises key metrics about each school, including:
  * School Name
  * School Type
  * Total Students
  * Total School Budget
  * Per Student Budget
  * Average Maths Score
  * Average Reading Score
  * % Passing Maths
  * % Passing Reading
  * % Overall Passing (The percentage of students that passed maths **and** reading.)
  
* Create a dataframe to hold the above results

```python
#set index to school name 
school_type = school_data.set_index(["school_name"])["type"]

#creating separate dataframes for each of the conditions
#will use these dataframes to find the total of students who passed only maths, only reading and both
school_passing_maths = school_data_complete[(school_data_complete["maths_score"] >= 50)]
school_passing_reading = school_data_complete[(school_data_complete["reading_score"] >= 50)]
school_passing_maths_reading = school_data_complete[(school_data_complete["reading_score"] >= 50) & (school_data_complete["maths_score"] >= 50)]

#group original dataframe by school name
grouped_school_df = school_data_complete.groupby(["school_name"])

grouped_total_students=grouped_school_df["student_name"].count()
grouped_total_students
grouped_total_school_budget=grouped_school_df["budget"].mean()
# grouped_total_school_budget
grouped_per_student= grouped_total_school_budget/grouped_total_students
# grouped_per_student
grouped_average_maths = grouped_school_df["maths_score"].mean()
# grouped_average_maths_reading

grouped_average_reading = grouped_school_df["reading_score"].mean()

# percentage of passed maths = number of students who passed maths at each school/total 
# number of students at each school and time by 100
#we can do calculations with columns in different dataframes here because they're both grouped by school_name
#i.e., share the same index (i.e., school_name) and both have the same length
percentage_passed_maths = school_passing_maths.groupby(["school_name"]).count()["student_name"]/grouped_total_students*100
percentage_passed_maths


percentage_passed_reading = school_passing_reading.groupby(["school_name"]).count()["student_name"]/grouped_total_students*100
percentage_passed_reading

percentage_passed_overall = school_passing_maths_reading.groupby(["school_name"]).count()["student_name"]/grouped_total_students*100
percentage_passed_overall

#create a dataframe to hold all of the found values
school_summary= pd.DataFrame({"School Type": school_type,
                              "Total Students": grouped_total_students,
                              "Total School Budget": grouped_total_school_budget,
                              "Per Student Budget": grouped_per_student,
                              "Average Math Score": grouped_average_maths,
                              "Average Reading Score": grouped_average_reading,
                              "% Passing Math": percentage_passed_maths,
                              "% Passing Reading": percentage_passed_reading,
                              "% Overall Passing": percentage_passed_overall})
school_summary

#mapping for column Total School Budget 
school_summary["Total School Budget"] = school_summary["Total School Budget"].map("${:,}".format)
school_summary

#mapping for column Per Student Budget
school_summary["Per Student Budget"] = school_summary["Per Student Budget"].map("${:}".format)
school_summary
```

## Maths Scores by Year

* Create a table that lists the average maths score for students of each year level (9, 10, 11, 12) at each school.

  * Create a pandas series for each year. Hint: use a conditional statement.
  
  * Group each series by school
  
  * Combine the series into a dataframe
  
  * Optional: give the displayed data cleaner formatting

```python
#create a dataframe for year 9 values using loc[], group them by school_name and calculate the mean score for each school
year_9 = school_data_complete.loc[school_data_complete["year"]== 9].groupby(["school_name"]).mean()["maths_score"]
year_9
# #create a dataframe for year 10 values using loc[], group them by school_name and calculate the mean score for each school
year_10 = school_data_complete.loc[school_data_complete["year"]== 10].groupby(["school_name"]).mean()["maths_score"]
# year_10
# #create a dataframe for year 11 values using loc[], group them by school_name and calculate the mean score for each school
year_11 = school_data_complete.loc[school_data_complete["year"]== 11].groupby(["school_name"]).mean()["maths_score"]
# year_11 
# #create a dataframe for year 12 values using loc[], group them by school_name and calculate the mean score for each school
year_12 = school_data_complete.loc[school_data_complete["year"]== 12].groupby(["school_name"]).mean()["maths_score"]

#create a dataframe to hold new values
maths_scores_by_year = pd.DataFrame ({"Year 9": year_9,
                                  "Year 10": year_10,
                                  "Year 11": year_11,
                                  "Year 12": year_12})
maths_scores_by_year
```
